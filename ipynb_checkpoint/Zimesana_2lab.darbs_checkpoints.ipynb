```python

from numpy import *
x= linspace(0,7,70)
y= cos(x)
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funkcija $cos(x)$')
plt.plot(x, y, color = "#8fbc8f")#krāsas izvēle
plt.show()
```
